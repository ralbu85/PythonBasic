# 리스트

지금까지의 프로그램에서는 변수에 하나의 값만을 저장할 수 있었다. 하지만 프로그램에서 여러개의 값을 묶어서 동시에 처리해야 하는 경우가 매우 많다. 

예를 들어 50명이 수강하는 수업에서 학생의 성적 정보를 처리하는 프로그램을 만드려고 한다. 어떻게 해야할까?  지금까지 배운 방법으로는 100개의 서로 다른 이름을 가진 변수를 하나하나 선언해야만 했다. 수강생의 규모가 커질수록 이같은 방식은 매우 비효율적일 수밖에 없다. 

수학의 벡터(vector)처럼 프로그램에서도 여러개의 값을 묶어서 하나의 값으로 표현하는 체계가 존재한다. 파이썬에서는 이같은 값을 리스트(list)라고 한다. 

## 1. 리스트도 또 다른 형태의 값(value)이다.

<blockquote>

</blockquote>
<p>

리스트는 또 다른 형태의 값이기 때문에 변수에 저장할 수 있다.

## 3. 리스트 인덱싱 (리스트 값에 접근하기)

리스트에 여러개의 값이 저장되어 있으므로, 이중 특정 값에 접근하거나 변경하는 기능이 필요하다. 

리스트안에 들어있는 값은 모두 상대적인 주소를 가지고 있는데 이를 인덱스(index)라고 한다. 이 인덱스 값을 사용하여 리스트안의 특정 값에 접근하는 것을 인덱싱(indexing)이라고 한다. 다음과 같이 리스트 이름에 중괄호를 하고 이 안에 인덱스 값을 넣으면 된다.
<blockquote>
리스트 [인덱스]
</blockquote>
<p>

n개의 값을 갖는 리스트의 경우 0부터 n-1의 값을 가지며, 가장 왼쪽의 값을 0으로 한다. 다음 예시를 살펴보자.

```python
a=[1,3,5,7,9]
print(a[0]) #첫번째 값 (제일 왼쪽)
print(a[1]) #두번째 값
print(a[4]) #마지막 값 (제일 오른쪽)
```

인덱스를 사용하여 리스트 안에 들어있는 값을 바꿀수도 있다. 할당연산자 =과 인덱스를 이용해 리스트안의 특정위치에 있는 값을 바꾸면 된다. 만약 리스트 a에 첫번째 있는 값을 10으로 바꾸고 싶다면 다음과 같이 실행하면 될 것이다.

```python
a[0]=10
print(a)
```

인덱스에 마이너스를 붙여 오른쪽부터 접근할 수 도 있다. 다음 예시를 살펴보자

```python
print(a[-1]) #마지막 값(제일 오른쪽)
print(a[-2]) #오른쪽에서 두번째 값
print(a[-5]) #제일 왼쪽 값
```

리스트 안에는 리스트가 들어있을 경우에도 인덱스는 작동한다. 다음 예시를 살펴보자.

```python
a=[[1,2,3],[4,5],'life']
print(a[0])
```

a[0]이 리스트 [1,2,3]을 의미하므로 a[0][0]을 하면 리스트 [1,2,3]중 첫번째 값인 1이 나오게 된다.

```python
print(a[0][0]) # [1,2,3] 중 첫번째 값을 의미
print(a[1][1]) # [4,5] 중 두번째 값
```

## 4. 리스트 슬라이싱

지금까지 인덱스는 리스트 안에 들어있는 하나의 값을 추출하는데 사용했다. 
인덱싱 기능을 확장하여 리스트의 일부분을 추려내는 기능이 존재한다. 이를 리스트 슬라이싱(slicing)이라고 한다. 리스트 슬라이싱 문법은 다음과 같다.
<blockquote>
리스트 [시작인덱스:마지막인덱스]
</blockquote>
<p>
이를 실행하면 시작인덱스부터 마지막인덱스-1 번째 인덱스 값까지 연속된 리스트 값이 출력된다 (마지막 인덱스까지가 아님에 유의할 것). 
    
다음 예시를 살펴보자.    


```python
a=[1,3,5,7,9]
print(a[0:3]) #첫번째 값 (제일 왼쪽)
```

```python
print(a[:2]) #시작인덱스를 생략하면 처음값 0부터 시작한다
```

```python
print(a[2:]) #마지막인덱스를 생략하면 마지막 값까지 추려낸다.
```

## 5. for는 list와 함께 사용된다. 

파이썬에서 for와 list는 매우 빈번히 사용된다. Module 6에서 다루었던 for 문의 문법을 다시한번 살펴보겠다.
<blockquote>
for 변수 in range(지정횟수):<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
</blockquote>
<p>

사실 for문은 range대신 모든 리스트와 함께 사용이 가능하다. 따라서 이 문법을 조금더 엄밀하게 바꾸면 다음과 같이 표현할 수 있다.

<blockquote>
for 변수 in 리스트:<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
</blockquote>
<p>

리스트와 함께 사용된 for문의 예시를 살펴보자.

```python
test_list = ['one', 'two', 'three'] 
for i in test_list: 
    print(i)
```

<p> 눈여겨 봐야 할 부분은 for문에서 선언한 변수 i이다. for문은 리스트에 존재하는 값을 하나씩 참조하여 i라는 변수에 대입한다.  ['one', 'two', 'three']라는 리스트의 첫 번째 요소인 'one'이 먼저 i 변수에 대입된 후 print i라는 문장을 수행한다. 다음에 'two'라는 두 번째 요소가 i 변수에 대입된 후 print i 문장을 수행하고 리스트의 마지막 요소까지 이것을 반복한다. </p>

## 7. range의 정체

사실은 range 명령어는 리스트를 축약하여 생성할 수 있는 명령어이다. 다음 예시를 실행해보자.

```python
for i in range(5):
    print(i)
```

```python
my_list=[0,1,2,3,4]
for i in my_list:
    print(i)
```

위의 예시에서 알 수 있듯이 range(5)는 0부터 시작하여 4까지 연속된 값으로 구성된 리스트 [0,1,2,3,4]를 축약하여 만들어 놓은 표현이다. 따라서 for문은 리스트와 맞물려 작동한다고 표현하는 것이 정확하다고 하겠다.

range는 다양한 옵션이 있다. 다음의 예시를 살펴보자.

```python
for i in range(1,5): #1부터 시작하여 4까지 연속된 값으로 구성된 리스트 생성
    print(i)
```

```python
for i in range(0,10,2): #0부터 9이전까지 2씩 증가하면서 리스트 생성 즉 [0,2,4,6,8]
    print(i)
```

```python
# 중첩된 for문을 사용하여 구구단 1단부터 9단까지 출력하는 프로그램을 작성해보시오.
# 1*1=1
# 1*2=2
# ...
# 9*8=72
# 9*9=81
```

## 8. in 과 리스트

리스트 안에 특정한 값이 있는지 여부를 확인하는 작업도 프로그램에서 굉장히 많이사용된다. 이에 따라 파이썬에서는 in이라는 명령어를 제공하여 리스트안에 특정한 값이 있는지 확인하도록 도와주고 있다. 다음 예시를 살펴보자.

```python
mylist=[1,3,5,7,9]
3 in mylist
```

3 값이 mylist 안에 들어있기 때문에 true가 반환되었다. 만약 특정 값이 들어있지 않다면 다음처럼 false를 반환한다.

```python
4 in mylist
```
