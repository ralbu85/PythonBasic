# 반복문

프로그램을 실행하다보면 특정 처리문을 계속 반복해야 할 경우가 발생한다. 예를들어 사용자로부터 입력값을 받아 5번 출력하는 프로그램이 있다고 하자. 반복문을 배우지 않은 상태에서는 다음과 같이 할 코드를 작성할 수밖에 없다.


```python
variable = input('값을 입력하세요:')
print(variable)
print(variable)
print(variable)
print(variable)
print(variable)
```

하지만 반복의 횟수를 늘려 천번 반복해야 한다면 어쩔 것인가? ctrl c,v 신공을 사용하여 1000번 코드를 붙여 넣을 수 있지만, 이보다 효율적인 방법이 존재한다는것을 직감적으로 알 수 있을 것이다. 프로그램의 어떤 처리문을 반복하기 위한 명령어를 반복문이라고 한다. 파이썬에서는 크게 두 가지 종류의 반복문이 존재한다.

- for (지정된 횟수 반복)
- while (조건을 만족하는 한 계속해서 반복)

## 1. for (지정된 횟수 반복)

for 문은 지정된 횟수만큼 반복할 때 사용한다. 즉 프로그램 실행전, 5번, 100번 등 프로그래머가 미리 지정해 놓은 값만큼 특정 구문을 반복할때 사용할 수 있다. for문의 구조는 다음과 같다.

<blockquote>
for 변수 in range(지정횟수):<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
</blockquote>
<p>

첫번째 'for 변수 in range(지정횟수)' 부분은 반복의 횟수를 지정하는 부분이다. 변수명은 사용자 마음대로 정의해도 된다. 보통 i등의 간략한 알파벳으로 사용한다. (현재는 이정도로만 알아두자.) 사용자가 반복하고 싶은 횟수를 지정횟수부분에 숫자로 넣으면 된다. 

두번째 부분은 조건문과 마찬가지로 반복하고자 하는 구문들을 모아서 코드블럭처리한 것이다. 탭으로 띄어쓰기 된 코드들은 반복시 한꺼번에 실행될 것이다.

예를들어 위의 예제와 똑같은 프로그램을 for 문을 통해 다음과 같이 코딩할 수 있다.

```python
variable = input('값을 입력하세요:')
for i in range(5):
    print(variable)
```

## 2. 반복문 안에서 변수의 값 갱신하기

지금까지 예시에서는 반복문 안에서 특정한 값을 출력하는 예시를 다루었다. 
반복문은 같은 값을 출력할때 뿐만 아니라, 반복적으로 계산하면서 변수에 값을 동적으로 갱신하는 경우도 빈번히 발생한다. 다음의 예시를 살펴보자. 

우선 x에 0이라는 초기값을 저장했다. 

```python
x=0
```

그리고 다음 코드를 실행하여 x에 저장된 결과를 살펴보자

```python
x=x+3
print(x)
```

같은 코드를 다시한번 실행해보자.

```python
x=x+3
print(x)
```

위 코드를 반복적으로 수행할때 마다 x 변수에 3이 누적되어 더해지는 것을 알 수 있다. 
할당연산자 =를 중심으로 오른쪽 부분은 지금까지의 x값을 의미하고, 왼쪽은 새롭게 연산이 끝난 후 (3을 더한 뒤)의 값을 새로 갱신하는 과정이 반복되기 때문에, 코드를 실행할때마다 x변수에는 3이 누적되게 된다. 이 패턴은 알고리즘등의 연산에서 매우 빈번하게 활용되기 때문에 반드시 숙지할 필요성이 있다.

위에서 설명한 코드구조와 for문을 결합하여 반복된 횟수만큼 3을 더하여 출력하는 프로그램을 설계해 보면 다음과 같을 것이다.

```python
result=0
for i in range(5):
    result=result+3
print(result)
```

## 3. 중첩반복문

if else문을 중첩해서 사용가능했듯이, for문도 중첩해서 사용이 가능하다. 다음의 예시를 살펴보자.

```python
for i in range(2):
    print('hi')
    for j in range(3):
        print('there')
```

for가 두번 반복되어 복잡해 보이지만, 천천히 살펴보면 매우 쉽다. 첫째줄의 정의에 따라 이 코드는 코드블럭 2~4 line을 반복한다. 이중 3번째 line이 또다른 반복문으로 there를 세번 반복해서 출력한다. 첫번째 for문이 실행될 때마다 hi there there there이 실행되게 되고 이를 두번 실행했으므로 위와 같은 결과가 출력되는 것이다. 

## 4. while (조건을 만족하는 한 계속 반복)

지정된 횟수만큼 반복하는 for문과 달리, while문은 조건을 만족하는한 끊임없이 실행문을 반복한다. while문의 구조는 다음과 같다.

<blockquote>
while 조건문:<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
</blockquote>
<p>

'while 조건문:' 조건문 부분은 if에서 사용된 조건문과 동일하다. while문에서는 프로그램을 반복할때마다 조건문을 체크해보고, 현재 조건이 True라면 계속 프로그램을 실행하게 된다. 1부터 10까지 값을 출력하는 프로그램을 while로 구현한다면 다음과 같이 구현할 수 있을 것이다.

```python
count=1 #출력할 변수, 초기값 1로 설정
while count<=10: 
    print(count)
    count=count+1 # count에 1씩 누적시켜서 숫자 증가시키기
```

이 예제에서 조건문은 count<=10, 즉 count 변수에 있는 값이 10 미만이면 계속해서 코드블럭 부분을 반복해서 실행하게 된다. 4번째 줄에서 반복이 실행될 때마다 count의 값이 1씩 증가하게 된다. 이 코드가 반복되어 count가 10이 넘어가는 순간 조건을 만족하지 않게 되고, 더이상 반복문을 실행하지 않게 되는 것이다.
