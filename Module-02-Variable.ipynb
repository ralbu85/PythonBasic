# 변수(variable)

앞서, 프로그램을 값을 처리하기 위한 성능좋은 계산기로 정의했다.<br>
값을 처리하는 연산과정이 복잡해지면, 중요한 의미를 갖는 값을 어딘가에 저장했다가 나중에 다시 꺼내서 활용해야할 경우가 굉장히 많이 있다.<br>
<br>
프로그램에서 값을 저장했다가 다시 꺼내어 사용하기 위한 그릇으로 변수(variable)라는 것이 존재한다.

## 0. 변수에는 값을 저장해야 한다.

변수를 만들기 위해서는 거창한 작업이 필요하지 않다. 그냥 독자 마음대로 이름을 붙이는 과정이 전부이다. 예를들어 원주율 값을 저장하기 위한 변수로 pi라는 이름의 변수를 만들고자 한다면, 그대로 pi라는 이름으로 프로그램 안에서 사용하면 된다. 
<br>
컴퓨터가 이를 인식하는지 살펴보자.

```python
pi
```

지금 상태에서는 에러메시지가 출력되는데, 이는 변수에 값이 저장되어 있지 않기 때문이다.<br>
즉, 변수안에 값을 저장해야만 프로그램에서는 변수를 사용가능한 것으로 인식한다.<br>

## 1. 변수에 값을 저장하기 위해서는 =를 사용한다.

변수에 값을 저장하는 방법은 간단하다. <br>
다음과 같이 왼쪽에는 변수를 놓고 오른쪽에 여러분이 저장할 값을 넣으면 된다.
- 변수 = 값

예를들어 pi라는 변수에 3.141592라는 값을 저장하고 싶다면 다음과 같이 하면 된다.

```python
pi=3.141592
```

이제 에러메시지 없이, 컴퓨터가 pi라고 하는 변수에 3.141592라는 값이 저장되었음을 인식하였다.

## 2. 변수에 저장된 값 확인

이제 변수에 이름만 호출해도 프로그램은 변수안에 저장되어 있는 값을 바로 확인할 수 있다. <br>
변수에 저장된 값을 확인하기 위해서는 다음과같이 변수의 이름만 입력해도 된다.

```python
pi
```

## 3. 변수는 변할수 있다.

변수라는 단어에서 알수 있듯이, 변수안에 저장된 값은 변할 수 있다. <br>
현재 pi라고하는 변수에는 3.141592라는 값이 담겨 있다. <br>
이제 마음이 변해서 3.141592 대신 100이라는 값을 변수에 저장하고 싶다면 현재 상태에서 그대로 새로운 값으로 덮어씌우면 된다.

```python
pi=100
```

이제 pi에는 3.141592대신에 새로운 값인 100이 저장되었다.

```python
pi
```

## 4. 변수에는 어떤형태의 값도 저장할수 있다.

변수 안에는 어떠한 형태의 값도 저장할 수 있다.

```python
pi='hello world'
pi
```

## 5. 변수명에는 규칙이 있다.

따라서 프로그램의 관리를 위해 의미를 가지도록 변수명을 정하는 것이 매우 중요하다. 복잡한 프로그램에는 100개 이상의 변수가 사용되는데 변수명을 의미 없이 붙인다면 프로그램이 어떤 일을 하는지 이해하기 어려울 뿐만아니라, 유지관리에도 상당한 애를 먹을 수 있다.

또한 파이썬에서는 변수명에 관한 규칙이 있는데 다음과 같다.
- 변수명은 숫자부터 시작할 수 없다
- 변수에 _를 제외한 특수문자를 사용할 수 없다.
