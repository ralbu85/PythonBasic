# 조건문

## 0. 조건에 따른 프로그램의 분기

지금까지 살펴본 프로그램은 모두 직선적인 구조를 가지고 있었다. 프로그램이 여러줄로 구성되어 있다 하더라도, 순서대로 처리문을 실행시켜 나가는 구조 였다는 것이다. 
다음의 프로그램을 생각해보자.
<blockquote> 숫자를 입력받아 my_number라는 변수에 저장하고, 여기에 100을 곱하여 출력한다. </blockquote>
이를 프로그래밍하면 다음과 같이 순서대로 코드를 진행하면 될 것 이다.

```python
my_number = input("숫자를 입력하세요.")
my_number = int(my_number)
print(my_number*100)
```

하지만 프로그램이 복잡해지면 직선적인 처리로 불가능한 상황이 많이 발생하게 된다. 대표적으로 조건에 따라서 서로 다른 처리문을 실행해야 하는 경우가 바로 그것이다. 예를들어 다음 프로그램을 살펴보자.
<blockquote> 이 프로그램은 마찬가지로 숫자를 입력받아 my_number에 저장하지만, 사용자가 10보다 큰 수를 입력했을 경우에는 -100을 곱하고, 10보다 작은 수를 입력했을 경우에는 100을 곱하여 출력한다. </blockquote>
    
이 프로그램은 조건에 따라 서로 다른 2개의 처리문이 실행되어야 한다.
만약 입력된 값을 my_number라는 변수에 저장했다고 했을때, 
10보다 작은 경우에는 my_number * 100을 실행해야 하고, 10보다 큰 경우에는 number * -100를 실행해야 한다는 말이다. 현재로써는 이를 수행할 방법이 존재하지 않는다. 프로그램에서 조건에 따라 프로그램의 처리를 분기하기 위해서는 조건문이 필요하며 if와 else라는 명령문을 이용한다.

## 1. if문의 기본구조
다음은 if와 else를 이용한 조건문의 기본 구조이다.
<blockquote>
if 조건문:<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
else:<br>
&emsp;수행문장1<br>
&emsp;수행문장2<br>
</blockquote>
<p>
<q>if 조건문:</q> 부분은 만족해야할 조건을 쓰는 공간이다. 조건을 테스트하여 True 값이면 실행하고 False 값이면 실행하지 않는다. (참고: 지난시간에 비교연산의 결과로 True나 False 두개의 값을 갖는 bool형 변수가 생성된다고 했었다.)<p> 다음 예제를 살펴보자. </p>    


```python
my_number = input("숫자를 입력하세요")
my_number = int(my_number)
if my_number>10:
    print(my_number*-100)
else:
    print(my_number*100)
```

세번째 줄인 my_number>10는 my_number에 저장된 변수가 10이상인지 물어보는 코드이며, 이에 대한 결과로 boolean 값을 반환하게 된다. 만약 True이면 첫번째 코드블럭이 실행되어 if 다음의 코드블럭이 출력될 것이고, 그렇지 않다면 else부분의 코드 블럭이 실행될 것이다.

이를 통해 지금까지 직선적이었던 프로그램의 구조를 조건에 따라 흐름을 바꾸어 실행할 수 있게 된 것이다.

## 1. 코드블럭

조건을 만족했을때 실행해야하는 처리문이 2줄 이상일 수 있다. 이를 구분하기 위해 조건을 만족시 함께 실행되는 부분을 묶어서 코드블럭을 형성하기 위해서는 탭키로 들여쓰기(indentation)을 해야한다. 다음 예시를 살펴보자.

```python
age = input("Your age?:")
myage = int(age)
if myage>30:
    print('아재입니다.')
    print('노력하세요')
else:
    print('곧 아재가 됩니다.')
    print('축하합니다.')
```

## 2. 중첩된 조건문

위의 예제 프로그램을 살펴보자. 현재 프로그램은 30살 이상이면 아재이고, 30살 미만이면 다른 메시지가 출력된다. 
조건을 더 세분화해서 30살 이상이면 아재, 20~30 사이에 있으면 청년, 20대 미만이면 청소년을 출력하고 싶다면 어떻게 해야할까?

우선 현재 상황에서 30살 미만인 경우, 조건문을 한번 더 추가하여 다음과 같이 실행하면 될 것이다. 

```python
age = input("Your age?:")
myage = int(age)
if myage>30: # 30이상인경우
    print('아재입니다.')
else:
    if myage>20: # 30미만이지만 20 이상인 경우
        print('청년입니다.')
    else:
        print('청소년입니다') #20 미만인 경우
        
        
```

## 3. elif를 사용한 코드단순화

위처럼 여러개의 조건이 있다 하더라도 if else 구문을 중첩하여 조건에 따라 프로그램을 실행할 수 있다. 하지만 비교해야하는 조건이 3개 이상이면 코드의 구조가 복잡해져 가독성이 떨어지게 된다. 이같은 문제를 해결하기 위해 파이썬에서는 else와 if를 결합하여 elif문을 제공한다.

```python
age = input("Your age?:")
myage = int(age)
if myage>30: # 30이상인경우
    print('아재입니다.')
elif myage>20: # 30미만이지만 20 이상인 경우
    print('청년입니다.')
else:
    print('청소년입니다') #20 미만인 경우
        
```

보다시피, 예제코드에서 5번째 줄의 else와 6번째 줄의 if를 결합하여 elif로 쓰고, 코드블럭또한 한단위로 줄어들어들게 된다. 위 코드보다 훨씬 가독성이 좋을 뿐만 아니라 이해도 빠르다.
