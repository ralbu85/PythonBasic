## 5. 문자열 대입하기

```python
"I eat {} apples".format(3)
```

```python
"I eat {} apples and {} bananas".format(2,1)
```

```python
"{} + {} = {}".format(3.0,2.3,5.3)
```

```python
"Select * from mydb where id={}".format('\'jihwan\'')
```

```python
name="jihwan"
```

```python
f"My name is {name}"
```

```python
var1='mydb'
id=30
f"Select {var1} from mydb where id={id}"
```

```python
var1='mydb'
id=30
f"Select {var1+'_1'} from mydb where id={id+30}"
```
