## 1. List Comprehension (looping)

* 아래패턴: 리스트에 있는 값을 처리해서 새로운 리스트에 추가하는 패턴
* 빈리스트를 만들고 for문을 사용해서 하나하나 추가

```python
numbers=[1,2,3,4,5]
doubled=[]

for i in numbers:
    doubled.append(i*2)
    
print(doubled)
```

* 이제 아래와 같이 리스트 컴프리헨션(list comprehension) 기법을 적용해보자.

```python
numbers=[1,2,3,4,5]
doubled=[i*2 for i in numbers]
doubled
```

* for문을 리스트 []에서 사용했다. 이를 통해 for와 append를 동시에 한줄로 축약하였다.

## 2. List Comprehension (filtering)

```python
numbers=range(10)
odd=[]

for i in numbers:
    if i%2==1:
        odd.append(i)
    
print(odd)
```

```python
odd=[i for i in numbers if i%2==1]
odd
```

```python
numbers=range(10)
odd=[]
for i in numbers:
    if i%2==1:
        odd.append('odd')
    else:
        odd.append('even')
odd
```

```python
numbers=range(10)
odd=['odd' if i%2==1 else 'even' for i in numbers]
odd
```

## 3. zip

```python
x=[1,2,3]
y=['cat','lion','dog']
zip(x,y)
```

```python
for i,j in zip(x,y):
    print(i,j)
    print('*'*10)    
```

## 4. enumerate

```python
for i,name in enumerate(['cat','lion','dog','pig']):
    print(i,name)
```

```python
for i,name in enumerate(['cat','lion','dog','pig']):
    if i>1:
        print(name)
```

```python
[name for i,name in enumerate(['cat','lion','dog','pig']) if i>1]
```
