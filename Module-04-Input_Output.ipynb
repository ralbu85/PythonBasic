# 출력과 입력 (print, input)

## 1. 명시적 출력(print)

파이썬에서는 특정한 일을 수행하는 명령문들이 정의되어 있다.이중에서 print는 매우 자주 사용되는 명령문으로써, 값이나 변수값을 명시적으로 출력하는데 사용된다.

print후 괄호안에 값이나 변수를 넣으면 해당 값이 출력된다.

```python
print(1)
```

```python
print('hello world')
```

```python
x=5
print(x)
```

```python
y=5
print(y)
print(y**10)
```

jupyter notebook 환경에서 이미 입력한 값이나 변수를 넣으면 그대로 출력이 되는데 굳이 왜 print함수가 필요할까? 그 이유는 여러 실행문으로 구성된 프로그램에서 내가 명시적으로 변수나 값을 선택하여 출력할 수 있기 때문이다.

다음 예시를 통해 살펴보자.

```python
x=1
y=2
z=x+y
z+3
```

위 코드를 살펴보면 4개의 실행문으로 구성되어 있으며, 마지막 실행문이 '암시적'으로 실행되어 6이라는 결과가 출력되었다.

현재 상태에서 세번째 실행문인 z에 저장된 값은 살펴볼 수 없다는 말이 된다. 만약 프로그램이 실행되면서 z에 저장된 값을 확인하고 싶다면 다음과 같이 print를 사용하면 된다.

```python
x=1
y=2
z=x+y
print(z) # z에 저장된 3 값이 출력된다.
z+3 # z+3의 값이 출력된다.
```

이제 z에 저장되어 있는 값이 먼저 출력되고 마지막 줄인 z+3의 결과가 실행되는 것을 확인할 수 있다.

```python
x=1
y=2
z=x+y
print(z) # z에 저장된 3 값이 출력된다.
print(z+3) # z+3의 값이 출력된다.
```

print문은 출력값을 확인할 때 뿐만 아니라, 연산 중간과정에서 저장된 변수가 올바르게 작동하는지 테스트 하는 용도로도 사용된다.

## 2. print의 사용예시

- 콤마로 묶어서 여러개의 값을 동시에 출력할수 있다.

```python
x=1
y=2
print(x,y)
print(x+1,y**10)
```

- 문자열과 섞어서 사용할 수 있다.

```python
x=1
y=2
print('result of x+y is', x+y)
```

```python
x=1
y=2
print('result of x+y is', x+y, 'Thank you very much!')
```

## 3. 사용자 지정 입력

프로그램 실행 중 사용자가 직접 값을 입력하는 경우가 많이 있다.

## 4. 값 형태변환

어떤 프로그램에서 input함수를 사용하여 사용자가 입력한 숫자를 변수로 저장하고 싶다.
다음과 같은 프로그램을 실행하여 number에 숫자를 저장해보면 문제가 발생한다.

```python
number=input('Enter the number please:')
```

```python
number
```

즉 number에 저장된 값은 숫자가 아니라 문자임을 알수 있다. 숫자형태이지만 문자형태로 저장이 되었다는 말이다. 이를 숫자형태로 파이썬에게 인식시키기 위해서는 형태변환 명령어를 사용해야 한다.

int/float 명령어는 문자열 값이 숫자일 경우 이를 각각 정수와 소수형으로 변환시켜주는 함수이다. 사용자가 숫자를 입력했기때문에 number에 다음과 같이 적용할 수 있다.

```python
int(number) #저장된 값을 정수형 숫자로 변환
```

```python
float(number) #저장된 값을 소수형 숫자로 변환
```
