## 파이썬스런 코딩(Pythonic Coding)

## 1. lambda

```python
add = lambda x:x+1
add(3)
```

```python
# 어떤 값에 제곱하는 함수를 lambda 함수로 만들어 보시오
```

```python
add = lambda x,y:x+y
add(3,4)
```

## 2. map

```python
result=[]
for i in range(10):
    result.append(i*2)
result
```

```python
mysum=lambda x:x*2
mysum(10)
```

```python
list(map(mysum,range(10)))
```

```python
list(map(lambda x:x*2,range(10)))
```

```python
# 문자열의 갯수를 반환하는 함수를 map과 lambda로 작성해보시오.
# ['cat','lion','dog','elephant'] -> [3,4,3,8]
```

```python
check=lambda x:1 if x>5 else 0
print(check(10))
print(check(4))
```

```python
list(map(check,range(10)))
```

## 3. filter

```python
numbers=[-5,-3,0,1,3]
result=[]
for i in numbers:
    if i>0:
        result.append(i)
result
```

```python
cond = lambda x:x>0
print(cond(5))
print(cond(-5))
```

```python
list(filter(cond,numbers))
```

```python
list(filter(lambda x:x>0, numbers))
```

```python
#연습문제
#복수조건 필터를 수행하시오.
```

```python
numbers=range(10)
list(map(lambda x: x*2, (filter(lambda x:x>5,numbers))))
```

```python
numbers=range(10)
list(filter(lambda x: x>5, (map(lambda x:x*2,numbers))))
```

```python
numbers=range(1,11)
reduce(lambda x,y:x+y, numbers)
```
