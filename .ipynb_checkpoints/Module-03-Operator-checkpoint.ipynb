# 연산자(operator)

프로그램은 값(value)을 연산해가면서 원하는 값을 출력해가는 과정이라고 정의했다.<br>
값의 연산을 위해서는 연산자(operator)라고 하는 기호가 필요하다. <br>

## 0. 사칙연산

기본적인 연산자로(operator)의 숫자의 사칙연산을 들 수 있다.
- +(더하기)
- -(빼기)
- *(곱하기)
- /(나누기)

위 연산자들은 숫자와 숫자사이에 붙어서 사칙연산을 수행한다.
다음의 예시를 통해 살펴보자.

```python
1 + 1 #덧셈
```

```python
10-2 #뺄셈
```

```python
2 * 3.141592 #곱셈
```

```python
-3/2 #나눗셈
```

위의 예시에서 살펴볼수 있듯이, 두 값에 연산자를 적용한 결과, 새로운 값이 계산되어 출력되는 것을 볼 수 있다. 연산자는 이처럼 기존의 값을 새로운 값으로 변환시키는 프로그램의 아주 중요한 요소이다.

## 1. 변수와 값 사이의 연산

변수이름만으로 변수에 저장된 값을 불러올수 있었다. <br>
따라서 변수사이에서도 연산자가 적용된다.<br>
예를들어 x라는 변수에 5를 저장하고 y라고 하는 변수에 7을 저장했다면, 다음과 같이 사칙연산이 모두 가능하다. 

```python
x=5
y=7
```

```python
x+y
```

```python
x-y
```

```python
x*y
```

```python
x/y
```

마찬가지로 변수와 값 사이에서의 연산도 가능하다.

```python
x+5
```

```python
y/5
```

## 2. 연산의 결과를 변수에 저장

위에서 살펴보았듯이 연산의 결과 값이 출력되는데<br>
연산의 결과 또한 값(value)이므로, 변수에 저장할수 있다.
변수에 값을 저장할때와 마찬가지로 연산의 결과값을 저장할 변수를 왼쪽에 놓고 =를 사용하면
연산의 결과를 변수에 저장할 수 있게 된다. 

```python
result = 5 + 5
result
```

```python
result = 5 / 2
result
```

## 3. 사칙연산 외 연산자

사칙연산 외에도 다양한 연산자들이 존재한다. 다음의 예시를 통해 살펴보자.

```python
## 제곱근 (**)
5**2
```

```python
## 나머지 (%)
5 % 2
```

## 4. 값의 형태가 달라지면 연산자의 의미도 달라진다.

값의 형태가 달라지면 연산자의 의미도 달라진다.<br>
문자열과 문자열 사이에서는 +를 사용할 수 있는데 어떤 결과가 출력되는지 다음의 예시를 살펴보자.

```python
'hello' + 'world' # 문자열 사이에서 +는 두 문자열을 붙이는 연산자이다.
```

```python
'hello' * 3 # 문자열뒤에 * 숫자값이 들어오면 숫자 값만큼 반복해서 문자열을 붙이게 된다.
```

## 5. 비교연산자

비교연산자는 컴퓨터에게 질문을 던져 답이 참인지 거짓인지 물어볼수 있는 연산자이다.<br>
예를들어 3이 5보다 큰가? 이같은 질문에 우리는 거짓이라고 분명하게 답할수 있다.<br>
컴퓨터도 마찬가지이다. 컴퓨터에게 우리가 참/거짓으로 분명하게 답할 수 있는 질문을 던진다면 컴퓨터는 이를 판단하여 참(True)/거짓(False) 여부를 알려준다. <br>
앞서 다루었던 불리언 값(boolean value: True, False)은 이러한 참 거짓 여부를 나타내는데 사용되게 된다.
<br>

그렇다면 어떤 질문에 참거짓으로 명확한 대답이 가능할까?
크게 두 가지 종류의 질문으로 나누어 볼 수 있다.
각각의 질문에 대응되는 연산자는 괄호로 표시해 놓았다.

- 두 값중 어떤 한 값이 큰가? (대소 비교 : >, <, >=, <=)
- 두 값이 같은 값인가? (같은지 비교 : ==)
- 두 값이 다른 값인가? (다른지 비교: !=)

다음의 예시를 살펴보자. 대소 비교하는 연산자의 경우 왼쪽부터 오른쪽으로 읽어나가면 편하다.

```python
3 < 5 # 3이 5보다 작은가?
```

```python
3 >= 5 # 3이 5보다 크거나 같은가?
```

마찬가지로 두 값이 같은지 물어보기 위해서는 == 연산자를 활용하면 된다.

```python
3 == 3
```

```python
3 == 5
```

```python
3 != 5
```

당연히 비교연산자는 변수 사이에서도 사용할 수 있다.

```python
x = 5
y = 3
x < y
```

```python
x == y
```

또한 비교 연산자의 결과를 다른 변수에 저장할 수 있다.

```python
z = x < y
z
```

```python
z = x==y # 연습문제
z
```

## 6. 논리연산자

논리연산자는 참/거짓(불리언 값) 값에 대해서만 적용할 수 있는 연산자이다.
다음 두개의 연산자로 구성되어 있다. 
 - and (두 불리언 값이 참일 경우에만 참을 리턴)
 - or (둘 중 하나의 값이 참일경우 참을 리턴)

여러분은 논리학, 집합등을 접하면서 and와 or의 쓰임새에 대하여 이미 알고 있을 것이다. 
<br> and 연산자의 경우 두 불리언 값이 참일 경우에만 참값을 반납한다.

```python
True and True
```

```python
False and True
```

```python
False and False
```

```python
35 >10 and 20 >10
```

```python
35 <=10 and 20 >10
```

```python
True or False
```

```python
35 <=10 or 20 >10
```

## 7. 연산자의 우선순위 (feat 괄호)

한 실행문에 여러개의 연산자가 존재한다면 어떻게 될까?<br>
파이썬에서는 다음의 연산자 우선순위 규칙에 따라 판단한다.

- 1. **
- 2. \* / %
- 3. \+ - 

```python
12 + 6 ** 2 / 2 # 1순위: 6**2=>36  2순위: 36/2=18, 3순위 12+18=30
```

하지만 괄호로 연산자를 묶을경우 괄호에 있는 연산이 가장 먼저 계산된다.<br>
여러가지 연산이 섞여있을 경우 괄호를 사용하여 명확하게 연산의 순서를 정해주는것을 권장한다. 

```python
12 + 6 ** (2/2) # 1순위: 2/2, 2순위: 6**1=6, 3순위: 12+6
```
