# 값(value)

값(value)은 프로그램의 기본단위이다.<br>
대부분의 컴퓨터 프로그램은 프로그램은 사용자로부터 입력값(input value)을 받아, <br>
적절한 연산(computation)을 통해 출력값(output)을 도출한다.<br>
<br>
아주 단순한 사칙연산부터 딥러닝(deep learning)과 같은 복잡한 프로그램까지,<br>
컴퓨터 프로그램은 모두 예외없이 값을 계산하는 절차를 구성해 놓은 것이라고 할 수 있겠다.<br>

## 0. 값의 종류

프로그램에서 다루는 다양한 값의 종류가 존재한다.<br>
파이썬을 비롯한 대부분의 프로그램에서 값은 크게 다음과 같이 세 종류로 나누어 볼 수 있다.
- 숫자
- 문자(문자열)
- 불리언(논리)

## 1. 숫자 (numeric value)

숫자(number)는 프로그램에서 처리할 수 있는 가장 기본적인 값이다.<br>
다음 셀을 실행해보자.

```python
1
```

에러없이 여러분이 입력한 1이라는 숫자값을 그대로 출력하는 것을 볼 수 있다.<br>
프로그램에서 여러분이 입력한 값을 이해하였다고 생각하면 된다.<br>
여러가지 숫자형식에 대하여 실행해보자

```python
-20001
```

정수 뿐만 아니라 소수 값도 표현이 가능하다.

```python
13.5
```

```python
3.141592
```

## 2. 문자 (string value)

프로그램이 인식할 수 있는 또 다른 형태의 값으로 문자가 존재한다.<br>
숫자외의 모든 값들은 문자라고 볼 수 있다. x라는 문자를 파이썬이 인식하는지 살펴보기 위해 다음과 같이 셀을 실행해 보자.

```python
x
```

에러가 발생하였을 것이다. 컴퓨터가 x라는 문자를 인식할 수 없다고 나온다.<br>
(정확히는 x가 정의되어 있지 않다고 말하고 있다.) <br>
여기서 알수 있듯이, 문자값을 컴퓨터에게 인식시키기 위해서는 다른 약속이 필요함을 알 수 있다.<br>
다음 셀 처럼 x를 작은따옴표 '로 감싸서 코드를 실행해보자.

```python
'x'
```

이제 프로그램이 x라는 문자를 인식하고 있음을 알 수 있다.<br>
프로그램에서 문자를 값으로 사용하고 싶다면 인용부호 ''를 사용하면 된다.<br>
큰따옴표 ""도 사용 가능하다

```python
"x" 
```

### 2.1 다양한 문자

```python
' ' #공백도 하나의 문자열이다
```

```python
'!' #특수문자도 하나의 문자열이다
```

```python
'\'' #' 자체를 문자열로 하고 싶다면 
```

### 2.2 큰따옴표와 작은따옴표 차이점

큰따옴표와 작은따옴표는 기능적으로 차이점이 없다. <br>
다만 큰따옴표 안에서는 작은따옴표를 문자로 사용할 수 있다.

```python
"'"
```

### 2.3 문자열

프로그램에서는 한글자의 문자 뿐만아니라 <br>
문장처럼 여러개의 문자로 이루어진 값도 처리할 수 있다.<br>
사용법은 간단하다. 여러 문자를 따옴표로 묶으면 된다.

```python
'hello world'
```

```python
'!@#_!!!!1111'
```

```python
"Professor's name is Jihwan Lee" #문자열 내에 작은따옴표가 사용되어 큰따옴표로 묶어서 처리했다.
```

```python
'Professor\'s name is Jihwan Lee'
```

## 3. 불리언 값 (boolean value)

위의 예시에서 문자열을 따옴표로 묶지않고 그냥 실행했을 경우 값으로 인식하지 않는다는 것을 확인했다.<br>
그런데 다음처럼 실행을 해보자.

```python
True
```

```python
False
```

위 두 값은 실행했을 경우 컴퓨터가 인식한다.<br>
파이썬에서는 특별하게 True/False 두가지 값에 대해 특별한 의미를 부여하였는데 <br>
숫자나 문자와는 다른 제3의 형태의 값이라고 생각하면 된다. <br>
문자 그대로 True는 참, False는 거짓을 의미한다.<br>
위 두 값은 프로그램의 흐름을 제어하는데 굉장히 중요한 역활을 하게 될 것이므로 반드시 알아두어야 한다.

## 4. type을 이용한 값의 종류 확인

컴퓨터 프로그램은 내가 입력한 값을 이해할 뿐만 아니라, <br>
이 값이 어떤종류인지까지도 자동으로 이해할 수 있다.<br>
이를 확인해보기 위해 다음의 명령어를 실행해보자. 

```python
type(1)
```

```python
type(3.141592)
```

```python
type('a')
```

```python
type('hello world')
```

```python
type(True)
```

```python
type(False)
```
