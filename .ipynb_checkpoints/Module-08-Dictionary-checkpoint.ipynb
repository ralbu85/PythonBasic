# Dictionary(딕셔너리)

사전의 구조를 살펴보면 다음과 같다.
- (python): 명사, 비단뱀, 이무기, 거대한 뱀
- (pyramids): 명사, 피라미드, 이집트 거대 구조물.

위에서 알 수 있듯이 사전은 어떤 단어와 그에 맞는 뜻이 한쌍으로 맞물려서 의미를 갖게 된다.
이처럼 key와 value의 쌍으로 이루어진 자료형을 딕셔너리(dictionary) 자료형이라고 하며, 다른 곳에서는 해쉬(hash)라는 단어로 사용되기도 한다. 

딕셔너리는 리스트나 튜플처럼 순차적으로(sequential) 해당 요소값을 구하지 않고 Key를 통해 Value를 얻는다. 이것이 바로 딕셔너리의 가장 큰 특징이다. baseball이라는 단어의 뜻을 찾기 위해 사전의 내용을 순차적으로 모두 검색하는 것이 아니라 baseball이라는 단어가 있는 곳만 펼쳐 보는 것이다.

## 1. 딕셔너리의 구조

딕셔너리의 구조는 다음과 같다.
- {key1:value1, key2:value2, key3:value3, ....}

key 값은 사전안의 키워드에 해당한다고 볼 수 있겠고, value는 key값을 설명하는 구체적인 값이라고 볼 수 있겠다. 리스트로 보면 key값은 인덱스에 가깝고, value는 리스트에 순차적으로 저장된 값이라고 볼 수 있겠다.

```python
{'name':'jihwan','age':23, 'height':183, 'weight': 67}
```

딕셔너리 또한 리스트와 처럼 하나의 값으로 인식된다. 따라서 변수에 저장할수 있다.

```python
person={'name':'jihwan','age':23, 'height':183, 'weight': 67}
print(person)
```

## 2. 딕셔너리에 저장된 값에 접근

리스트에서 인덱스를 [ ]안에 사용하듯이, 딕셔너리에서는 키값을 [ ]안에 사용하여 원하는 value를 호출할 수 있다. 위에서 정의한 person 딕셔너리의 예시를 살펴보겠다.

```python
person['name']
```

```python
person['age']
```

리스트처럼 딕셔너리에 저장된 값도 키값과 할당연산자 =를 사용하여 바꿀수 있다. 이 사람의 나이를 33으로 바꾸고 싶다면 다음처럼 하면 될 것이다.

```python
person['age']=33
```

```python
print(person)
```

이처럼 키값을 활용하여 리스트의 인덱스처럼 활용할 수 있다.

## 3. 다차원 딕셔너리

리스트와 마찬가지로 딕셔너리값은 형태에 제한이 없다. 특히 딕셔너리 값으로 딕셔너리를 가질수 있는데, 이러한 다차원 딕셔너리는 실제로 굉장히 많이 사용된다. 

다음 예시에서 person 딕셔너리는 키값으로 이제 사람의 이름을 가지고 있으며, 각 키에 해당하는 값에는 그 사람의 신상정보를 또다른 딕셔너리가 담겨있다. 이 딕셔너리에서 'jihwan'키를 선택하면 다음과 같은 jihwan에 대한 신상정보가 담긴 딕셔너리가 출력될 것이다.

```python
person={'jihwan':{'age':23, 'height':183, 'weight': 67},
        'bogum':{'age':25,'height':185,'weight':63},
       'woosung':{'age':40,'height':187,'weight':70}}
person['jihwan']
```

jihwan의 몸무게 정보를 보고싶다면 다음과 같이 하면 될 것이다. 

```python
person['jihwan']['weight']
```

woosung의 나이를 보고싶다면 다음과 같이 하면 될것이다.

```python
person['woosung']['age']
```

위에서 정의한 person 딕셔너리의 경우 age, hegiht, weight 세개의 키로 정의된 동일한 딕셔너리를 값으로 갖고 있는데, 이는 웹의 데이터 교환양식으로 널리사용되는 json(JavaScript Object Notation)과 같은 형태이다.

## 4. for와 딕셔너리 

이전 모듈에서 for문과 리스트를 사용하여 리스트에 있는 값들에 하나씩 접근이 가능하였다. 
마찬가지로 for문과 딕셔너리르 결합하여 딕셔너리에 저장된 정보에 순차적으로 접근할수 있다.
다음 예시를 살펴보자.

```python
person={'name':'jihwan','age':23, 'height':183, 'weight': 67}
for i in person:
    print(i)
```

for문에서 순차적으로 접근하는 변수 i를 살펴보니 딕셔너리의 키값에 하나씩 접근하는 것을 볼 수 있다. 이를 활용하여 딕셔너리에 저장된 값에도 하나씩 접근이 가능하다. 다음의 예시를 살펴보자.


```python
for i in person:
    print(i, person[i])
```

두번째 줄에서 print(i, person[i])를 하게되면 키값과 딕셔너리의 value값을 쌍으로 출력할 수 있게 된다.
