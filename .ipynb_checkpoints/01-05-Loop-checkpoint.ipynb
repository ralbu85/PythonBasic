## 반복가능한 객체: 리스트

````python
[1,35, 7] # 여러개의 값을 중괄호 안에 넣으면 리스트가 된다.
````

````python
x = [1, 35, 7, 20] 
y = ['a', 35, 'hello', 20] # 다양한 종류의 값을 담을 수 있다
z = [True, False, 1, 3] # Boolean값도 담을 수 있다
m = [] #아무값도 없어도 리스트다
print(x)
print(y)
print(z)
print(m)
````

````python
mixed = ['hello',[1,2],[]] #리스트 안에 리스트도 담을 수 있다
print(mixed)
````

## for를 이용한 반복

````python
# 리스트는 반복가능한 객체이므로 for와 함께 사용할 수 있다.
numbers = [1,2,3,4,5]
for i in numbers: # 리스트 안에 값을 순서대로 꺼내어 i에 대입
    print(i)
````

````python
# 동일한 작업을 지정된 횟수만큼 반복하기
numbers = [1,2,3,4,5] # 지정할 횟수만큼의 갯수를 가진 리스트 생성
for i in numbers: 
    print('hello') #반복할 코드
````

````python
# 리스트 내부의 값을 순서대로 활용한 코드
for i in numbers:
    print(i+3) # 각 값에 3씩 더함
````

````python
temp = 'any'
print('Hi' + temp + 'Man!')
````

````python
x = ['kim','na','park','lee']
for i in x:
    print('Hi', i, 'Nice to meet you!')
````

## 5-3. range 명령어

````python
# 반복해야하는 숫자만큼 일일히 리스트의 값을 넣어주는 것은 비효율적임
# 자동으로 리스트를 만들어주는 range 함수를 많이 사용함
range(5) # [0,1,2,3,4]와 똑같음
````

````python
for i in range(5):
    print(i)
````

````python
for i in range(5):
    print('hello world')
````

````python
for i in range(5):
    print(i**2)
````

## 6-5. 중첩반복문

````python
for i in range(2):
    print('hi')
    for j in range(3):
        print('there')
````

````python
for i in [1,3,5]:
    for j in ['a','b','c']:
        print(i,j)
````

## 6-6. while 반복문

````python
while True:
    print("hey")
````

````python
while False:
    print("hey")
````

````python
i=0
while i<5:
    i=i+1
    print(i)
````

## 6-7. continue

````python
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
````

````python
for val in "Data":
    if val == "a":
        continue
    print(val)

print("The end")
````

````python
var = 5
while var > 0:
    var = var - 1
    if var == 2:
        continue
    print('Current variable value :', var)
print("Good bye!")
````

## 6-8. break

````python
for i in range(10):    # 0부터 9999까지 반복
    print(i)
    if i == 3:    # i가 100일 때
        break       # 반복문을 끝냄. for의 제어흐름을 벗어남
````

````python
i = 0
while True:    # 무한 루프
    print(i)
    i += 1          # i를 1씩 증가시킴
    if i == 12:    # i가 100일 때
        break       # 반복문을 끝냄. while의 제어흐름을 벗어남
````

````python
answer = 3
while True:
    x = int(input("입력해보세요:"))
    if(x==answer):
        print("맞았습니다.")
        break
    print("틀렸습니다.")
````
