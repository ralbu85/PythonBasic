# 함수(function)

프로그래밍을 하다 보면 같은 내용을 반복해서 작성하고 있는 자신을 발견할 것이다. 이때가 바로 함수가 필요한 부분이다. 함수란 정해진 어떤 일을 수행하기 위한 명명되어진(이름이 붙은) 코드의 묶음이라고 볼 수 있다.

함수는 다음과 같은 특성을 가지고 있다.
- 변수와 마찬가지로 이름이 붙어져 있다.
- 반복적으로 사용되는 가치 있는 부분을 한 뭉치로 묶는다.
- 어떤 입력값을 주었을때 어떤 결과값을 돌려받게 만든다.


## 1. 함수의 정의
<p> 함수는 다음과 같이 정의한다.</p>
<blockquote>
    def 함수명():<br>    
    &emsp;명령문1<br>
    &emsp;명령문2<br>
</blockquote>


- 첫번째 줄은 함수를 선언하는 부분으로 def 뒤에 함수의 이름을 붙인다. 변수와 마찬가지로, 파이썬에서 미리 지정해놓은 몇가지 명령문을 제외하고 마음대로 함수의 이름을 지정할 수 있다.
- 함수선언 후에는 코드블럭이 따라온다. 함수가 호출되었을때 실행해야할 명령문의 집합이 이 부분에 정의된다.

예를들어 다음 코드의 두 줄이 프로그램에서 반복적으로 사용된다고 하자.

```python
print('Hello')
print('World')
```

위의 반복적으로 실행되는 부분을 묶어서 print_twice라는 이름의 함수로 정의하고자 하면 다음처럼 정의하면 된다.

```python
def print_twice():
    print ('Hello')
    print ('Hello')
```

## 2. 함수의 호출

위 코드를 실행했을때에는 아무 일도 일어나지 않는다. 현재까지는 함수를 정의만 했을 뿐 호출하지 않았기 때문이다. 여러분이 정의한 함수를 실제로 사용하기 위해서는 함수를 호출하는 과정이 필요하다.

함수를 호출하기 위해서는 정의된 함수의 이름을 사용하면 된다. 위에서 정의내린 print_twice의 경우 다음처럼 호출한다. (참고: 주피터 노트북 환경에서는 위의 셀에서 실행된 결과를 그대로 간직하고 있음을 기억하자.)

```python
print_twice()
```

## 3. 함수의 입력값(인자, parameter)

위에서 정의한 print_twice함수는 hello라는 문자열을 두번 연속으로 출력하는 함수였다. 만약에 상황에 따라서 hello라는 문자열 대신에 world라는 단어를 두번 출력하고 싶다면 어떻게 해야할까? 다른이름의 함수를 정의하기에는 시간과 공간이 너무 아깝다. 이럴때에는 함수에 입력값(인자)를 정해줌으로써 손쉽게 해결할 수 있다.

<p> 함수에 입력값을 지정해 주기 위해서는 함수명 뒤에 붙는 괄호안에 상황에 따라 다르게 사용될 입력값을 변수로 지정해준다. </p>
<blockquote>
    def 함수명(입력값 변수):<br>    
    &emsp;명령문1<br>
    &emsp;명령문2<br>
</blockquote>

이를 바탕으로 print_twice 함수를 바꾸어 다음과 같이 정의할 수 있다.

```python
def print_twice(my_string):
    print(my_string)
    print(my_string)
```

print('hello')였던 이전 함수와 달리 my_string이라는 변수가 사용되고 있고, my_string은 함수의 입력값 부분에 정의되어 있다.

이제 함수를 호출하면서 입력값을 지정해주면 주어진 입력값에 따라 정의된 함수의 출력내용이 바뀌는 것을 볼 수 있다.

```python
print_twice('world')
```

```python
print_twice('lovely twice')
```

```python
print_twice(123)
```

함수 호출시 사용자가 입력한 값이 my_string이라는 변수에 저장되고, 이에 따라 정의된 코드블럭이 실행된다. 따라서 사용자가 서로 다른 값을 함수에 입력하면 이에따라 다른값이 출력되게 되는 것이다.

## 4. 입력변수는 함수 안에서만 유효하다.

위에서 정의한 print_twice함수의 입력값으로 my_string 변수를 사용했다. my_string에 현재 어떤 값이 저장되어 있는지 확인해보기 위해 다음 코드를 실행해보면 에러가 나는 것을 확인할 수 있을 것이다. 

```python
print(my_string)
```

이는 함수 안에서 지정된 변수는 함수안에서만 유효할뿐 함수가 호출되는 부분에서 사용할수 없기 때문이다. 함수정의시 사용된 변수는 함수 밖에서 사용할 수 없음에 유의하자.

## 5. 함수의 출력값

함수에서 수행한 연산 결과를 재사용하기 위해서는 어떻게 해야할까? 예를들어 섭씨온도를 입력받아 화씨온도를 계산하는 함수가 있다고 하자. 프로그래머는 이 함수를 통해 얻어진 화씨온도를 변수에 저장하여 또 다른 명령문에서 사용하고자 할 것이다. 

함수에서 얻어진 결과값을 재사용하기 위해 출력값을 정의해 주어야 한다. 출력값의 정의는 다음과 같이 한다.

<blockquote>
    def 함수명(입력값 변수):<br>    
    &emsp;명령문1<br>
    &emsp;명령문2<br>
    &emsp;return 출력 변수
</blockquote>

위와 같이 출력하고자 하는 변수 앞에 return 명령어를 넣어주면 된다. 
다음 예시를 살펴보자.

```python
def temperature(celc):
    ferh = celc * 1.8 + 32
    print(ferh)
```

첫번째 줄에서 celc를 입력변수로 정의하고,
두번째 줄에서 celc를 변환하여 ferh라는 변수에 저장하였으며
print를 통해 연산의 결과를 출력하고 있다.

```python
temperature(32)
```

```python
result=temperature(32)
print(result)
```

함수를 실행했을 경우 print 명령문으로 인해 화면으로 결과값이 출력되지만 이 값을 프로그램의 다른 부분에서 재사용할 수 없다. reseult 변수에 아무값도 저장되지 않았음에 유의하라. (None은 아무값도 저장되어있지 않다는 뜻이다.) 
return 명령어를 이용하여 출력변수를 지정해주면 다음과 출력값을 프로그램의 다른 부분에서 사용할 수 있게 된다.

```python
def temperature(celc):
    ferh = celc * 1.8 + 32
    return ferh
```

```python
result=temperature(32)
print(result)
```
